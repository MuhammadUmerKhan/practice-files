```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(color_codes=True)
```

```python
data = pd.read_csv('./laptop_price - dataset.csv')
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data['Company'].value_counts()
sns.barplot(x=data['Company'].value_counts().index, y=data['Company'].value_counts().values)
plt.xticks(rotation = 40)
plt.xlabel("Company")
plt.ylabel("Counts")
plt.title("Most Popular Companies")
plt.show()
```

```python
companies_revenue = data.groupby('Company')['Price (Euro)'].sum().sort_values(ascending=False).reset_index()
sns.barplot(data=companies_revenue, x='Company', y='Price (Euro)')
plt.xlabel('Companies')
plt.xticks(rotation = 45)
plt.ylabel("Revenue")
plt.title("Companies Total Revenue")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='Inches')
plt.title('Distribution of Inches')
plt.xlabel('Inches')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 10))
sns.histplot(data['ScreenResolution'], kde=True)
plt.title('ScreenResolution')
plt.xlabel('ScreenResolution')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(data['CPU_Company'].value_counts(),labels=data['CPU_Company'].value_counts().index, autopct='%1.1f%%')
plt.title('CPU_Company Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.histplot(data['CPU_Frequency (GHz)'])
plt.title('CPU_Frequency (GHz)')
plt.xlabel('CPU_Frequency (GHz)')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.histplot(data['RAM (GB)'])
plt.title('RAM (GB)')
plt.xlabel('RAM (GB)')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.histplot(data['OpSys'])
plt.title('OpSys')
plt.xlabel('OpSys')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```
